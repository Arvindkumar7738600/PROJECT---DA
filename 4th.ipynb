```python

# Making a new Series and DataFrame to store the Job Title with its frequency
unique_job_titles = pd.Series(df["Job Title"].value_counts())
unique_job_titles_df = unique_job_titles.to_frame(name='Unique Job Titles')
```

```python
#This will help us display all the rows
pd.set_option('display.max_rows', None)  
# This will list all the Job Titles which have a frequency of greater then 3 
unique_job_titles_df[unique_job_titles_df["Unique Job Titles"] >3].sum()
```
